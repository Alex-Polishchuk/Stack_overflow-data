```python
#Split each row of LanguageWorkedWithColumn into individual programming languages & creating a new dataframe
split_data = df["LanguageHaveWorkedWith"].str.split(";", expand=True)
```

```python
#Checking new df head to see if correct output
split_data.head()
```

```python
#Check which data, see how many responses per column
split_data.isnull().sum()
```

```python
rows_to_drop = []  # List to store indices of rows to be dropped

# Iterate through rows and identify rows with NaN values in the first column
for i, row in split_data.iterrows():
    cell_val = row.iloc[0]  # Access the value of the first column
    if pd.isna(cell_val):
        rows_to_drop.append(i)

# Drop the identified rows
split_data.drop(rows_to_drop, inplace=True)
```

```python
split_data.head()
```

```python
split_data.isnull().sum()
```

```python
#Removing columns 20 --> due to unlikeliness of devs using this many tools and being proficient.
for i in range (20, 42, 1):
    split_data.drop(i, axis=1, inplace=True)
```

```python
split_data
```

```python
df_series = split_data.stack()
```

```python
df_series = df_series.value_counts()
```

```python
#Exporting the data to a csv
df_series.to_csv("Programming_Dev_Use.csv")
```

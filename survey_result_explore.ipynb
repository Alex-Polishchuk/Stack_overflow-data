```python
#Split each row of LanguageWorkedWithColumn into individual programming languages & creating a new dataframe
LanguageHaveWorkWith = df["LanguageHaveWorkedWith"].str.split(";", expand=True)
```

```python
#Checking new df head to see if correct output
LanguageHaveWorkWith.head()
```

```python
#Check which data, see how many responses per column
LanguageHaveWorkWith.isnull().sum()
```

```python
rows_to_drop = []  # List to store indices of rows to be dropped

# Iterate through rows and identify rows with NaN values in the first column
for i, row in LanguageHaveWorkWith.iterrows():
    cell_val = row.iloc[0]  # Access the value of the first column
    if pd.isna(cell_val):
        rows_to_drop.append(i)

# Drop the identified rows
LanguageHaveWorkWith.drop(rows_to_drop, inplace=True)
```

```python
LanguageHaveWorkWith.head()
```

```python
LanguageHaveWorkWith.isnull().sum()
```

```python
#Removing columns 20 --> due to unlikeliness of devs using this many tools and being proficient.
for i in range (20, 42, 1):
    LanguageHaveWorkWith.drop(i, axis=1, inplace=True)
```

```python
LanguageHaveWorkWith
```

```python
LanguageHaveWorkWith_series = LanguageHaveWorkWith.stack()
```

```python
LanguageHaveWorkWith_series = LanguageHaveWorkWith_series.value_counts()
```

```python
#Exporting the data to a csv
LanguageHaveWorkWith_series.to_csv("Programming_Dev_Use.csv")
```

Calculating the fraction of people who use / want to learn a technology

```python
#List of all variables id need, number corresponds to the number of people who answered the q on survey
LanguageHaveWorkWith_ans = 70845
LanguageWantToWork_ans = 64737
DatabaseHaveWorked_ans = 59473
DatabaseWantToWork_ans = 50407
```

```python
#Convert all the series to a dataframe
LanguageWantToWork = LanguageWantToWork_series.to_frame()
LanguageHaveWorkWith = LanguageHaveWorkWith_series.to_frame()
DatabaseHaveWorked = DatabaseHaveWorked_series.to_frame()
DatabaseWantToWork = DatabaseWantToWork_series.to_frame()
```

```python
fractions = []
for index, row in LanguageHaveWorkWith.iterrows():
    value = row[0]
    fraction = value / LanguageHaveWorkWith_ans
    fractions.append(fraction)
LanguageHaveWorkWith['Fraction_of_people'] = fractions
```

```python
LanguageHaveWorkWith
```

```python
fractions = []
for index, row in LanguageWantToWork.iterrows():
    value = row[0]
    fraction = value / LanguageWantToWork_ans
    fractions.append(fraction)
LanguageWantToWork['Fraction_of_people'] = fractions
```

```python
fractions = []
for index, row in DatabaseWantToWork.iterrows():
    value = row[0]
    fraction = value / DatabaseWantToWork_ans
    fractions.append(fraction)
DatabaseWantToWork['Fraction_of_people'] = fractions
```

```python
fractions = []
for index, row in DatabaseHaveWorked.iterrows():
    value = row[0]
    fraction = value / DatabaseHaveWorked_ans
    fractions.append(fraction)
DatabaseHaveWorked['Fraction_of_people'] = fractions
```
